```python
import pandas as pd
import requests
from tqdm import tqdm
import glob
import os
```

```python
data_list = []
id = 575

for i in tqdm(range(1, 73)):
    url = "courses api url"
    response = requests.get(url)
    
    if response.ok:
        data = response.json()
        data_list.append(data)
    else:
        print(f"Request failed with status code {response.status_code}")
        
        
        
courses_data = []

for data in data_list:
    courses_data.extend(data['unit']['items'])
    
    
courses = pd.DataFrame(courses_data)
courses
```

```python
courses.to_csv('devtools_courses_data.csv', sep=';', index=False)
```
